```python
def get_businesses(query, api_key, min_rating, max_results):
    """Returns businesses nearby search location"""

    import requests
    import json

    query = query.replace(' ','+')
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=%s&key=%s" %(query, api_key)
    response = requests.get(url)

    response_dict = response.json()

    output = []
    interim = []
    name_list = []
    address_list = []
    hours_list = []
    prices_list = []
    rating_list = []
    
    for i in response_dict['results']:
        name_list.append(i['name'])
    for i in response_dict['results']:
        address_list.append(i['formatted_address'])
    for i in response_dict['results']:
        try:
            hours_list.append(i['opening_hours'])
        except:
            hours_list.append(0) 
    for i in response_dict['results']:
        try:
            prices_list.append(i['price_level'])
        except:
            prices_list.append(0)    
    for i in response_dict['results']:
        try:
            rating_list.append(i['rating'])
        except:
            rating_list.append(0) 
    interim = list(zip(name_list,address_list,hours_list,prices_list,rating_list))
    
    for item in interim:
        if item[-1]>=min_rating:
            output.append(item)
    
    if interim == output:
        interim = []
        name_list = []
        address_list = []
        hours_list = []
        prices_list = []
        rating_list = []
    
    while len(response_dict['next_page_token']) > 0 and max_results > len(output):
        try:
            npt = response_dict['next_page_token']
            new_url = 'https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&key=%s' %(npt, api_key)
            response = requests.get(new_url)
            while response.json().get('status')!='OK':
                from time import sleep
                from random import random
                sleep(random())
                response = requests.get(new_url)
            
            for i in response_dict['results']:
                name_list.append(i['name'])
            for i in response_dict['results']:
                address_list.append(i['formatted_address'])
            for i in response_dict['results']:
                try:
                    hours_list.append(i['opening_hours'])
                except:
                    hours_list.append(0) 
            for i in response_dict['results']:
                try:
                    prices_list.append(i['price_level'])
                except:
                    prices_list.append(0)    
            for i in response_dict['results']:
                try:
                    rating_list.append(i['rating'])
                except:
                    rating_list.append(0) 
            interim = list(zip(name_list,address_list,hours_list,prices_list,rating_list))
    
            for item in interim:
                if item[-1]>=min_rating:
                    output.append(item)
    
            if interim == output:
                interim = []
                name_list = []
                address_list = []
                hours_list = []
                prices_list = []
                rating_list = []
                
        except:
            print('end of data')
    print(len(output))
    return output
```
